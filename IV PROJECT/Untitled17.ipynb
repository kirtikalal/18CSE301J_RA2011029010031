```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Educational_status_of_suicide_victimes_state_2014_3.csv')
```

```python
data.drop(data.index[(data["States/UTs"] == "Total (UTs)")],axis=0,inplace=True)
data.drop(data.index[(data["States/UTs"] == "Total (States)")],axis=0,inplace=True)
data.drop(data.index[(data["States/UTs"] == "Total (All India)")],axis=0,inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=data, x='States/UTs', y='Total', hue='Year', palette='Set2', ax=ax)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.title('Total Number of Suicide in State')
plt.xlabel('States/UTs')
plt.ylabel('Total')
plt.xticks(rotation=90)

plt.show()
```

```python
data = pd.read_csv('Suicides_in_India.csv')
data.head()
```

```python
data_pie = data['Age_group'].value_counts().rename_axis('Age_group').reset_index(name='Total')
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie.Total, labels=data_pie.Age_group, startangle=90, autopct='%.1f%%')
plt.title('The different age suicides')
plt.show()
```
